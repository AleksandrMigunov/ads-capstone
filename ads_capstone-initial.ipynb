<h2>Data Case and Data Set</h2>

The use case for this project is analysis of the progress of vaccination against COVID in the world and new cases of COVID. The data set is Data on COVID-19 (coronavirus) by Our World in Data (https://github.com/owid/covid-19-data/tree/master/public/data) which contains data for every country and for the whole world taken from official sources. The goals of my project are: 1) to analyze how the progress of vaccination reduces the number of new cases of COVID; 2) to make a prognosis when the whole population of the world will be vaccinated.
